```python
import plotly

from plotly.offline import iplot, init_notebook_mode
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.io as pio

import os
import numpy as np
import pandas as pd
```

```python
def load_evals(u_inf):
    route = './cases/swf_u_inf%04d/swf_u_inf%04d/stability' % (u_inf*100, u_inf*100)
    evals_data = pd.read_csv(route + '/eigenvalues.csv')
    evals = evals_data['0'] + 1j*evals_data['1']
    return evals
```

```python
u_inf_range = np.array([10, 15, 20, 25, 28.75])
```

```python
evals = []
for i in u_inf_range:
    evals.append(load_evals(i))
```

```python
fig = go.Figure()


for it in range(len(u_inf_range)):
    fig.add_scatter(x=evals[it].real,
                y=evals[it].imag, 
               mode='markers');
iplot(fig)
    
```
