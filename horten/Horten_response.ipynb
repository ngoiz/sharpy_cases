# Wake Analysis

Evaluate difference in aerodynamic resultant forces between a horseshoe vortex model and a discrete wake model

```python
wake_data = pd.read_csv("csv_data/WakeConvergence.csv")

wake_data.groupby('m')
```

```python
A = ~np.isinf(wake_data['WakeLength'])
```

```python
B = wake_data['m']==4
C = wake_data['m']==6
wk4 = wake_data[A & B]
wk6 = wake_data[A & C]
```

```python
fig, ax = plt.subplots(2,1, figsize = (12,10), sharex = True)

ax[0].plot(wk4['WakeLength'],100*wk4['Rel_Fz'], 
        label = 'm = 4',
        lw = 2)
ax[1].plot(wk4['WakeLength'], 100*wk4['Rel_Fx'],
        label = 'm = 4',
        lw = 2)
ax[0].plot(wk6['WakeLength'], 100*wk6['Rel_Fz'],
        label = 'm = 6',
        lw = 2)
ax[1].plot(wk6['WakeLength'], 100*wk6['Rel_Fx'],
        label = 'm = 6',
        lw = 2)

ax[0].legend()
ax[1].legend()

ax[0].set_ylabel('F_z Percentage Delta')
ax[1].set_ylabel('F_x Percentage Delta')

ax[1].set_xlabel('Wake Length [m]')

plt.savefig('/home/ng213/Documents/PhD/WeeklyMeetings/Wk6_181031/wake_study.eps', dpi = 200)
```

# Effects of Inertia

Running with j = 0.5mu. This one has the same dt. Beware for following because dt changes with the 1/2 scale. output_j05


`data_j005.csv` j=0.05mu and dt = 1/4/30*0.5

Now running with inertia 5j and dt = 1/4/30*0.5

```python
def make_plot(csv_file,dt):
    data = pd.read_csv(csv_file)
    data['Time'] *= dt
    
    fig, ax = plt.subplots(2,1,figsize = (12,10), sharex = True)
    
    ax[0].plot(data['Time'], data['Z (stats)'], lw = 2, label = 'Z')
    ax[0].plot(data['Time'], data['X (stats)'], lw = 2, label = 'X')
    
    ax[0].legend()
    ax[0].set_ylabel('Displacement, [m]')
    
    theta = -np.arctan(data['local_y (2) (stats)']/data['local_y (0) (stats)'])*180/np.pi
    
    ax[1].plot(data['Time'], theta, lw = 2)
    
    ax[1].set_ylabel('Pitch Angle, \\theta [deg]')
    
    ax[1].set_xlabel('Time, t [s]')
    
    return fig, ax, theta
```

```python
def make_theta_plot(csv_file,dt):
    data = pd.read_csv(csv_file)
    data['Time'] *= dt
    
    fig = plt.figure(figsize = (12,10))
       
    theta = -np.arctan(data['local_y (2) (stats)']/data['local_y (0) (stats)'])*180/np.pi
    
    plt.plot(data['Time'], theta, lw = 2)
    
    plt.ylabel('Pitch Angle, \\theta [deg]')
    
    plt.xlabel('Time, t [s]')
    
    plt.grid()
    
    t = data['Time']
           
    return fig, theta, t
```

```python
fig, theta05, t05 = make_theta_plot('csv_data/data_j05.csv', 1/4/30)
plt.xlim(0,3)
```

```python
fig, ax, theta = make_plot('csv_data/data_j05.csv', 1/4/30)
fig.suptitle('Inertia j = 0.5 mu')
ax[1].set_xlim(0,3)
ax[1].grid()
```

SPPO frequency approx 3Hz, oscillations die out after 4 cycles/1.5s

```python
fig, ax, theta = make_plot('csv_data/data_j005.csv', 1/4/30/2)
fig.suptitle('Inertia j = 0.05 mu')

ax[1].set_xlim((0,2))
ax[1].grid()
```
